```python
import seaborn as sns
```

```python
##basisc ploting with seaborn

tips=sns.load_dataset('tips')
tips
```

```python
##crate the scatter plot
import matplotlib.pyplot as plt

sns.scatterplot(x='total_bill',y='tip',data =tips)
aaa =plt.title('sactter plot bill')
plt.show()
```

```python
##LINE PLOT

sns.lineplot(x='size',y='total_bill', data= tips)
plt.title("line polt")
plt.show()
```

```python
#catogiral polt
##bar pot
sns.barplot(x='day',y='total_bill', data =tips)
plt.title('VARIABLE')
plt.show()
```

```python
#box plot

sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
#voilin polt
sns.violinplot(x='day',y='total_bill', data = tips)
```

```python
#hsotogram
sns.histplot(tips['total_bill'], bins =10,kde=True)
```

```python
##KDE plot
sns.kdeplot(tips['total_bill'], shade = False, fill =True)
```

```python
#pairpolt
sns.pairplot(tips)
```

```python
#heatmap
cor =tips[['total_bill','tip','size']].corr()
cor
```

```python
sns.heatmap(cor,annot=True,cmap='coolwarm')
```

```python
import pandas as pd
sales_df = pd.read_csv('data.csv')
sales_df.head()
```

```python
# polt total sale by product

plt.figure(figsize=(10,6))
sns.barplot(x='Product',y='Sales', data = sales_df, estimator=sum )
plt.title('product sale')
plt.xlabel('product')
plt.ylabel('Sales')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Region',y='Sales', data = sales_df, estimator=sum ,color='pink')
plt.title('product sale')
plt.xlabel('product')
plt.ylabel('Sales')
plt.show()
```
