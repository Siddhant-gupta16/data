```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y =[1,4,9,16,25]

plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y lable')
plt.title("basic line plot")
plt.show()
```
